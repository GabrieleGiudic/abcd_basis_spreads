## ABCD parameterization of simple basis

```python
def add_forward_helpers(data, euribor):
    data['fwd_helper'] = None

    depos = data.ix['deposit']
    depos.loc[:,'fwd_helper'] = [ DepositForwardHelper(row.rate, euribor)
                                  for row in depos.itertuples() ]
    
    if 'fra' in data.index:
        fras = data.ix['fra']
        fras.loc[:,'fwd_helper'] = [ FraForwardHelper(row.rate, int(row.start), euribor)
                                 for row in fras.itertuples() ]
    
    if 'futures' in data.index:
        futures = data.ix['futures']
        futures.loc[:,'fwd_helper'] = [ FuturesForwardHelper(100.0*(1.0-row.rate), IMM.date(row.imm), euribor)
                                        for row in futures.itertuples() ]
    
    swaps = data.ix['swap']
    swaps.loc[:,'fwd_helper'] = [ SwapForwardHelper(row.rate, Period(row.tenor), calendar,
                                                    Annual, ModifiedFollowing,
                                                    Thirty360(Thirty360.BondBasis) if Period(row.tenor) >= Period(2,Years)
                                                    else Actual360(),
                                                    euribor, QuoteHandle(), Period(0, Days),
                                                    YieldTermStructureHandle(ois_curve))
                                  for row in swaps.itertuples() ]

add_forward_helpers(data6m, euribor6m)
```

```python
def simple_basis(index, base_index, guess, data):
    is_simple_basis = True
    basis = AbcdTenorBasis(index, base_index, spot,
                           is_simple_basis, guess)
    optimizer = LevenbergMarquardt(1e-8, 1e-8, 1e-8)
    helpers = list(data['fwd_helper'][data['calibrate']])
    basis.forwardCalibrate(helpers, optimizer)
    return basis

eonia = Eonia(YieldTermStructureHandle(ois_curve))
basis_6m = simple_basis(euribor6m, eonia,
                        [0.001624, 0.001026, 0.195197, 0.000193],
                        data6m)
```

```python
ABCD_curve_6m = TenorBasisForwardRateCurve(basis_6m)
ABCD_curve_6m.enableExtrapolation()
```

```python
euribor6m_ABCD = forward_index(Euribor6M(), ForwardRateCurveHandle(ABCD_curve_6m))
```

```python
def plot_ABCD_vs_legacy(euribor, legacy_euribor, legacy_curve, base_euribor):
    months = 12/euribor.tenor().frequency()
    dates = [ calendar.adjust(spot + Period(i, Months)) for i in range(months, 12*60) ]
    nodes, _ = zip(*legacy_curve.nodes()[1:])
    dates = sorted(set(dates+list(nodes)))
    legacy_basis = []
    ABCD_basis = []
    for d in dates:
        r_base = base_euribor.fixing(fixingDate(base_euribor, d))
        r_ABCD = euribor.fixing(fixingDate(euribor, d))
        s_ABCD = r_ABCD - r_base
        if d in nodes:
            r_x = legacy_euribor.fixing(fixingDate(legacy_euribor, d))
            s_x = r_x - r_base
        else:
            s_x = None
        ABCD_basis.append(s_ABCD)
        legacy_basis.append(s_x)
    utils.plot_curve(plt, dates, [(legacy_basis,'o'), (ABCD_basis,'-r')])

plot_ABCD_vs_legacy(euribor6m_ABCD, euribor6m, legacy_curve_6m, euribor6m_ois)
```

## Exact fit to market data

```python
def depo_rate(euribor):
    return euribor.fixing(today)

def fra_rate(months, euribor):
    start = TARGET().advance(spot, months, Months,
                             euribor.businessDayConvention(), euribor.endOfMonth())
    fixing_date = euribor.fixingDate(start)
    return euribor.fixing(fixing_date)

def futures_rate(imm, euribor):
    start = IMM.date(imm)
    fixing_date = euribor.fixingDate(start)
    return euribor.fixing(fixing_date)

def swap_rate(tenor, euribor):
    maturity = spot + Period(tenor)
    fixed_schedule = Schedule(spot, maturity, Period(1, Years), TARGET(),
                              ModifiedFollowing, ModifiedFollowing,
                              DateGeneration.Forward, False)
    floating_schedule = Schedule(spot, maturity, euribor.tenor(), TARGET(),
                                 euribor.businessDayConvention(), euribor.businessDayConvention(),
                                 DateGeneration.Forward, False)
    swap = VanillaSwap(VanillaSwap.Payer, 100000,
                       fixed_schedule, 0.0,
                       Thirty360(Thirty360.BondBasis) if Period(tenor) >= Period(2,Years)
                       else Actual360(),
                       floating_schedule, euribor, 0.0, euribor.dayCounter())
    swap.setPricingEngine(DiscountingSwapEngine(YieldTermStructureHandle(ois_curve)))
    return swap.fairRate()
```

```python
def add_implied_quotes(data, tag, euribor):
    data[tag] = None
    data.loc[:,tag].ix['deposit'] = [ depo_rate(euribor)
                                      for _ in data.ix['deposit'].itertuples() ]
    if 'fra' in data.index:
        data.loc[:,tag].ix['fra'] = [ fra_rate(int(row.start), euribor)
                                      for row in data.ix['fra'].itertuples() ]
    if 'futures' in data.index:
        data.loc[:,tag].ix['futures'] = [ futures_rate(row.imm, euribor)
                                          for row in data.ix['futures'].itertuples() ]
    data.loc[:,tag].ix['swap'] = [ swap_rate(row.tenor, euribor)
                                   for row in data.ix['swap'].itertuples() ]
```

## ABCD parameterization of continuous basis

```python
def continuous_basis(index, base_index, guess, data):
    is_simple_basis = False
    basis = AbcdTenorBasis(index, base_index, spot,
                           is_simple_basis, guess)
    optimizer = LevenbergMarquardt(1e-8, 1e-8, 1e-8)
    helpers = list(data['helper'][data['calibrate']])
    basis.calibrate(helpers, optimizer)
    return basis

cbasis_6m = continuous_basis(euribor6m, eonia,
                             [0.001624, 0.001026, 0.195197, 0.000193],
                             data6m)
```
